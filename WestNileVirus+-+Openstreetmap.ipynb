```python
import geopandas as gpd
from os import listdir
import time
import pandas as pd
from shapely.geometry import Point, Polygon, MultiPolygon, LinearRing
import numpy as np
```

```python
# Function definitions

# Function "log" makes a log with the added string, and mark it with the current time
def log(s):
    print('[' + time.strftime('%a %H:%M:%S') + '] ' + s)
```

```python
osmDir = 'C:/Users/kasina/Dropbox/Python tutorial/West Nile virus/map'

#put the shp filenames into a list
shapeFiles = [f for f in listdir(osmDir) if f.endswith('.shp')] 

#the 2nd and 3rd part of the name is the type of the shape file, let's put them into a list
shapeTypes = [(i.split("_")[3]) for i in shapeFiles] 

# Create a cellShapes GeoPandas DataFrame with reading all the cellshape files
cellShapes = None

log("Start reading shape files...")
for i in range(len(shapeFiles)):
    tmp = gpd.read_file(osmDir+"\\"+shapeFiles[i]) #if we use Lynux, we should use "/" instead of "\\"
    tmp['Shape_Type'] = shapeTypes[i]
    if cellShapes is not None:
        cellShapes = cellShapes.append(tmp)
    else:
        cellShapes = tmp
log("Finished reading shape files...")

cellShapes.head()
```
